```python
# Look at Log of values
num_of_points = 50
data_x = np.linspace(1, 25, num_of_points)       # Start, End and number of points
data_y = np.log(data_x)
# data_y = data_x * 3                            # Instead of log of values could be other function or multiple 
data_y
```

```python
plt.plot(data_x,data_y)  # Log function (Exponential behaviour)
```

### Adding labels to your plots

```python
plt.xlabel("Data Points")
plt.ylabel("Log of Data Points")
plt.plot(data_x,data_y) 
plt.show()
```

```python
# Look at Log of values
num_of_points = 100
data_x = np.linspace(1, 25, num_of_points)       # Start, End and number of points
data_ysin = np.sin(data_x)                       # Sin function
data_ycos = np.cos(data_x)

plt.plot(data_x,data_ysin) 
plt.plot(data_x,data_ycos) 

plt.show
```

### Specifying line attributes 
### Color : one of the parameters to plot function - 'b' (blue), 'g' (green), 'r' (red), 'm' (magenta), 'k' (black), 'c' (cyan), ' y' (yellow), 'w' (white)

### linewidth : specify a number for the width

### linestyle  : 'solid', 'dashed', 'dotted', 'dashdot'

```python
plt.plot(data_x,data_ysin, 'r', linewidth = 6, linestyle = 'dashed') 
plt.plot(data_x,data_ycos, 'b', linewidth = 2, linestyle = 'solid') 

plt.show
```

### Datapoints are referred to as markers
### You can specify type of market

```python
plt.plot(data_x,data_ysin, 'r', linewidth = 2, linestyle = 'dashed', marker = '*', markersize = 6) 
plt.plot(data_x,data_ycos, 'b', linewidth = 2, linestyle = 'solid', marker = 'o') 
plt.show
```
